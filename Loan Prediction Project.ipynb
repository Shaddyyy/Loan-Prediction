```python
df['Credit_History'] = df['Credit_History'].fillna(-1) #keeping the null values in the form of integer
```

```python
df['Credit_History'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset = 'Married', inplace = True)
```

```python
df['Married'].isnull().sum()
```

```python
df[df['Dependents'].isnull()] 
```

```python
df.dropna(subset = 'Gender', inplace = True)
```

```python
df['Dependents'] = df['Dependents'].fillna(4)
```

```python
df['Self_Employed'] = df['Self_Employed'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.head(2) #to see which columns need their values to be replaced or mapped, just for convinience
```

```python
df['Gender'] = df['Gender'].map({'Male':1,'Female':0}) #mapping the onject/string values to integer/float for passing in the logistic regression model
```

```python
df['Married'] = df['Married'].map({'Yes':1,'No':0})
```

```python
df['Dependents'] = df['Dependents'].map({'0':0,'1':1,'2':2,'3+':3})
```

```python
df['Self_Employed'] = df['Self_Employed'].map({'Yes':2,'No':1})
```

```python
df['Education'] = df['Education'].map({'Graduate':1,'Not Graduate':0})
```

```python
df['Property_Area'] = df['Property_Area'].map({'Urban':2,'Semiurban':1,'Rural':0})
```

```python
df['Loan_Status'] = df['Loan_Status'].map({'Y':1,'N':0})
```

```python
df.head() #value check
```

```python
df.corr()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics #for checking the accuracy
```

```python
df.columns #for selecting the independent variables
```

```python
x = df[['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed',
       'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area']]
y = df['Loan_Status']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.3, random_state = 101)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
predictions = lr.predict(xtest)
```

```python
metrics.accuracy_score(predictions,ytest)
```
